```python
from art import logo, vs
from game_data import data
from IPython.display import clear_output
from random import randint
from time import sleep
```

```python
def validatedInput(condition, prompt, failPrompt='INVALID INPUT', apply=None):
    while True:
        test = input(prompt)
        if condition(test):
            break
        else:
            print(failPrompt)
    return test if apply is None else apply(test)

class Game():
    
    @staticmethod
    def frmt(data):
        return f"{data['name']}, {'an' if data['description'][0].lower() in 'aeiou' else 'a'} "+\
    f"{data['description']}, from {data['country']}"
    
    def pick(self):
        picked = randint(0, self.max)
        
        if not hasattr(self, 'currentID'):
            self.currentID = -1
            
        if picked == self.currentID:
            self.pick()
        else:
            return picked
    
    def __init__(self, data):
        self.data = data
        self.max = len(self.data)-1
        self.currentID = self.pick()
        self.lost = False
        self.score = 0

    def refresh(self, texts):
        clear_output()
        print(logo)
        for text in texts:
            print('\n'*2, text)
        print(f"Score = {self.score}")

    def prompt(self):
        nextID = self.pick()
        
        self.data1 = self.data[self.currentID]
        self.data2 = self.data[nextID]
        
        compare = f"Compare: {Game.frmt(self.data1)}"
        against = f"Against: {Game.frmt(self.data2)}\n"
        self.refresh([compare, vs, against])
        
        prompt = 'Select Higher [H] or Lower [L]: '
        self.answer = validatedInput(lambda x: x.upper() in ['H', 'L'], prompt, apply=lambda x: x.upper())
        
        self.currentID = nextID
        
    def compare(self):
        diff = self.data1['follower_count'] - self.data2['follower_count']
        self.lost = (diff > 0 and self.answer == 'L') or (diff < 0 and self.answer == 'H')
        
        print(f"\nCounts: {self.data1['follower_count']} vs {self.data2['follower_count']}")
        
        if not self.lost:
            self.score += 1 
            print("Correct!")
            print(f"New Score = {self.score}")
            sleep(3)
        else:
            print("Wrong answer!")
        
    def play(self):
        
        while not self.lost:
            self.prompt()
            self.compare()
            
            
        print(f"Game Over")
```

```python
game = Game(data)
game.play()
```
